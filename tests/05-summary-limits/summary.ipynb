```python
import json
SETUP_PATH = '../common/setup.json'
f = open(SETUP_PATH)
data = json.load(f)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = ['.NET','Django','NestJS'] 
Y0 = [14469, 3159, 4132 ] 
Y100K = [90, 3070, 0] 
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, Y0, 0.4, label = 'FWB_0') 
plt.bar(X_axis + 0.2, Y100K, 0.4, label = 'FWB_100K') 

def addlabels(x,y, x_delta = 0):
    for i in range(len(x)):
        plt.text(i + x_delta, (y[i])//2, round(y[i], 2), ha = 'center', c="black")

addlabels(X_axis, Y0, x_delta = -0.2)
addlabels(X_axis, Y100K, x_delta = 0.2)
  
plt.xticks(X_axis, X) 
plt.ylabel("Liczba użytkowników") 
plt.title("Limity liczby użytkowników dla różnych wielkości danych") 
plt.legend() 
# plt.show() 
plt.savefig('Limit_vus_for_different_datasets.png')
```

```python
index = 0
fw = data["apps"][list(data["apps"].keys())[index]]
df = fw['df']

chart = sns.countplot(data=df, x='metric_name')
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
chart.bar_label(chart.containers[0])
chart.set_title(f'Quantity of records for {fw["name"]} with {data["vus"]} {"users" if data["vus"] > 1 else "user"}')
```
