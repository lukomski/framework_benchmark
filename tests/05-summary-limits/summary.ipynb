```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from unidecode import unidecode
```

```python
X = ['.NET','Django','NestJS'] 
Y0 = [10239, 3163, 131068 ] 
Y100K = [61, 2052, 165] 
  
X_axis = np.arange(len(X)) 

# fig1 = plt.figure(figsize=(10, 14), dpi=100)
  
plt.bar(X_axis - 0.2, Y0, 0.4, label = 'FWB_0') 
plt.bar(X_axis + 0.2, Y100K, 0.4, label = 'FWB_100K') 

def addlabels(x,y, x_delta = 0):
    for i in range(len(x)):
        plt.text(i + x_delta, (y[i])//2, round(y[i], 2), ha = 'center', c="black")

addlabels(X_axis, Y0, x_delta = -0.2)
addlabels(X_axis, Y100K, x_delta = 0.2)

plt.xticks(X_axis, X) 
plt.ylabel("Liczba użytkowników") 
plt.title("Limity liczby użytkowników dla różnych zbiorów") 
plt.legend() 
# plt.show() 
plt.savefig('Limit_vus_for_different_datasets.png')
```

```python

def draw_avg_limits(X: list[str], Y: list[float], title: str):
    
    X_axis = np.arange(len(X)) 
    
    plt.bar(X_axis, Y, 0.4) 

    def addlabels(x,y, x_delta = 0):
        for i in range(len(x)):
            plt.text(i + x_delta, (y[i])//2, round(y[i], 2), ha = 'center', c="black")

    addlabels(X_axis, Y, x_delta = 0)

    plt.xticks(X_axis, X) 
    plt.ylabel("Liczba użytkowników") 
    plt.title(title)
    plt.savefig(f'{unidecode(title.replace(" ", "_"))}.png')
```

```python
X = ['.NET','Django','NestJS'] 
Y0 = [10239, 3163, 131068 ] 

draw_avg_limits(
    X=X,
    Y=Y0,
    title=f'Limity liczby użytkowników dla zboru FWB_0'
    )
```

```python
X = ['.NET','Django','NestJS'] 
Y100K = [61, 2052, 165] 

draw_avg_limits(
    X=X,
    Y=Y100K,
    title=f'Limity liczby użytkowników dla zboru FWB_100K'
    )
```
