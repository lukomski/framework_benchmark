## Processing k6 results

```python
data = {
    'dotnet': {
        'url': 'http://localhost:5028/Member',
        'name': 'dotnet'
    },
    'django': {
        'url': 'http://localhost:8000/collisions',
        'name': 'django'
    },
    'nestjs': {
        'url': 'http://localhost:3000/member',
        'name': 'nestjs'
    },
}
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dir = 'out_2023-09-15_20-26-24'

for key in data:
    fw = data[key]
    path = f'./{dir}/{fw["name"]}.csv'
    print(path)
    data[key]['df'] = pd.read_csv(path)
```

```python
for key in data:
    fw = data[key]
    df = fw['df']
    df_1 = df.loc[(df['metric_name'] == 'http_req_duration')]
    sns.lineplot(data=df_1, x='timestamp', y='metric_value')
```

```python
for key in data:
    fw = data[key]
    df = fw['df']
    df_1 = df.loc[(df['metric_name'] == 'http_req_duration')]
    duration = df_1['metric_value']
    sns.lineplot(data=duration)
```
